```python
import seaborn as sns
import pandas as pd
from scipy.stats import chi2_contingency
import matplotlib.pyplot as plt

df = pd.read_csv('Life Expectancy Data.csv')
df.head()
```

```python
plt.figure(figsize=(16,5)) 
sns.heatmap(df.corr(), annot = True, vmin=-1, vmax=1, center= 0,linewidths=0.1)
```

```python
contingencyTable = df.groupby(['Status'])['Country'].size()
contingencyTable
```

```python
contingencyTable.plot(kind='bar')
```

# There is quite clearly a difference in categories here

```python
chi2_contingency(contingencyTable)
```
